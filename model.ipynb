```python
def clean_csv(df):
    # Remove commas, replace '-' with NaN, convert to float
    df['Volume']    = df['Volume'].astype(str).str.replace(',', '').str.strip().replace('-', np.nan).astype(float)
    df['Turn Over'] = df['Turn Over'].astype(str).str.replace(',', '').str.strip().replace('-', np.nan).astype(float)
    df['Percent Change'] = df['Percent Change'].astype(str).str.replace('%', '').str.strip().replace('-', np.nan).astype(float)
    return df

df_nabil       = clean_csv(df_nabil)
df_doorsanchar = clean_csv(df_doorsanchar)
df_citizen     = clean_csv(df_citizen)

print("✅ Data cleaned!")
```

```python
df_nabil.columns = df_nabil.columns.str.strip().str.upper().str.replace(" ", "_")

df_nabil['DATE'] = pd.to_datetime(df_nabil['DATE'])

df_nabil = df_nabil.sort_values('DATE')
```

```python
df_doorsanchar.columns = df_doorsanchar.columns.str.strip().str.upper().str.replace(" ", "_")

df_doorsanchar['DATE'] = pd.to_datetime(df_doorsanchar['DATE'])

df_doorsanchar = df_doorsanchar.sort_values('DATE')
```

```python
df_citizen.columns = df_citizen.columns.str.strip().str.upper().str.replace(" ", "_")

df_citizen['DATE'] = pd.to_datetime(df_citizen['DATE'])
df_citizen = df_citizen.sort_values('DATE')
```

```python
for df in [df_nabil, df_doorsanchar, df_citizen]:
    df['MA_5']       = df['CLOSE'].shift(1).rolling(5).mean()
    df['MA_10']      = df['CLOSE'].shift(1).rolling(10).mean()
    df['DAILY_RANGE']= df['HIGH'] - df['LOW']
    df['MOMENTUM']   = df['CLOSE'].shift(1) - df['CLOSE'].shift(6)
    df['PREV_HIGH']   = df['HIGH'].shift(1)  
    df['PREV_LOW']    = df['LOW'].shift(1) 
```

```python
df_nabil['PREV_CLOSE'] = df_nabil['CLOSE'].shift(1)
df_nabil = df_nabil.dropna() 

X_nabil = df_nabil[['PREV_CLOSE', 'MA_5', 'MA_10', 'DAILY_RANGE', 'MOMENTUM', 'PREV_HIGH', 'PREV_LOW']]
y_nabil = df_nabil['CLOSE']  
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# split data
X_train, X_test, y_train, y_test = train_test_split(X_nabil, y_nabil, test_size=0.2, shuffle=False)

# create model
model_nabil = LinearRegression()
model_nabil.fit(X_train, y_train)
y_pred = model_nabil.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
print("=== Nabil Bank ===")
print(f"MAE  : {mean_absolute_error(y_test, y_pred):.2f}")
print(f"RMSE : {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}")
print(f"R²   : {r2_score(y_test, y_pred):.4f}")

mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print(f"Accuracy : {100 - mape:.2f}%")
```

```python
plt.figure(figsize=(12, 4))
plt.plot(y_test.values, label='Actual', color='blue')
plt.plot(y_pred, label='Predicted', color='red', linestyle='--')
plt.title('Nabil Bank — Actual vs Predicted')
plt.legend()
plt.show()
```

```python
import joblib
import os

os.makedirs('models', exist_ok=True)

joblib.dump(model_nabil,       'models/model_nabil.pkl')
joblib.dump(model_doorsanchar, 'models/model_doorsanchar.pkl')
joblib.dump(model_citizen,     'models/model_citizen.pkl')

print("✅ All 3 models saved to models/ folder")
```

```python
from sklearn.linear_model import LinearRegression
import joblib, os
os.makedirs('models', exist_ok=True)

FEATURES = ['PREV_CLOSE', 'MA_5', 'MA_10', 'DAILY_RANGE', 'MOMENTUM', 'PREV_HIGH', 'PREV_LOW']

for df, name in [(df_nabil,'nabil'), (df_doorsanchar,'doorsanchar'), (df_citizen,'citizen')]:
    # High model
    m_high = LinearRegression()
    m_high.fit(df[FEATURES], df['HIGH'])
    joblib.dump(m_high, f'models/model_{name}_high.pkl')

    # Low model
    m_low = LinearRegression()
    m_low.fit(df[FEATURES], df['LOW'])
    joblib.dump(m_low, f'models/model_{name}_low.pkl')

print("✅ High and Low models saved!")
```
